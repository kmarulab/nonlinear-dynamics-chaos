```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the Rössler system
def rossler(t, state, a, b, c):
    x, y, z = state
    dxdt = -y - z
    dydt = x + a * y
    dzdt = b + z * (x - c)
    return [dxdt, dydt, dzdt]

# Parameters for the Rössler system
a = 0.2
b = 0.2
c = 5.7

# Initial conditions
initial_state = [1.0, 1.0, 1.0]

# Time range for simulation
t_start = 0.0
t_end = 100.0
t_points = 10000
t_eval = np.linspace(t_start, t_end, t_points)

# Solve the Rössler system
solution = solve_ivp(rossler, [t_start, t_end], initial_state, t_eval=t_eval, args=(a, b, c))

# Extract the solution
x, y, z = solution.y

# Plot the Rössler attractor
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_title("Rössler Attractor", fontsize=15)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```
