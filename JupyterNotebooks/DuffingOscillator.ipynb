```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the Duffing oscillator equations
def duffing(t, y, delta, alpha, beta, gamma, omega):
    x, v = y  # y[0] = x (position), y[1] = v (velocity)
    dxdt = v
    dvdt = -delta * v - alpha * x - beta * x**3 + gamma * np.cos(omega * t)
    return [dxdt, dvdt]

# Parameters
delta = 0.2   # Damping coefficient
alpha = -1.0  # Linear stiffness
beta = 1.0    # Nonlinear stiffness
gamma = 0.3   # Driving amplitude
omega = 1.2   # Driving frequency

# Initial conditions and time span
x0 = 0.5  # Initial position
v0 = 0.0  # Initial velocity
t_span = (0, 100)  # Time range
t_eval = np.linspace(t_span[0], t_span[1], 10000)  # Time points for output

# Solve the differential equation
sol = solve_ivp(duffing, t_span, [x0, v0], t_eval=t_eval, args=(delta, alpha, beta, gamma, omega))

# Extract position and velocity
x = sol.y[0]
v = sol.y[1]

# Plot the phase space (position vs velocity)
plt.figure(figsize=(8, 6))
plt.plot(x, v, lw=0.8)
plt.title("Duffing Oscillator Phase Space", fontsize=14)
plt.xlabel("Position (x)", fontsize=12)
plt.ylabel("Velocity (v)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
