```python
import matplotlib.pyplot as plt

# Define the Henon map
def henon_map(x, y, a, b):
    new_x = 1 - a * x**2 + y
    new_y = b * x
    return new_x, new_y

# Parameters for the Henon map
a_henon = 1.4
b_henon = 0.3

# Initial conditions
x_henon, y_henon = 0.1, 0.1

# Simulate the Henon map
iterations = 10000
henon_x = []
henon_y = []
for _ in range(iterations):
    x_henon, y_henon = henon_map(x_henon, y_henon, a_henon, b_henon)
    henon_x.append(x_henon)
    henon_y.append(y_henon)

# Plot the Henon map
plt.figure(figsize=(10, 6))
plt.plot(henon_x, henon_y, '.', markersize=0.5)
plt.title("Henon Map", fontsize=15)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```
